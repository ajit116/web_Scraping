# webscraping using python

http://www.fasttrack.co.uk/league-tables/tech-track-100/league-table/

```python
#dependencies
from bs4 import BeautifulSoup
from urllib.request import urlopen
import csv
```

```python
urlpage =  'http://www.fasttrack.co.uk/league-tables/tech-track-100/league-table/'
```

```python
souce = urlopen(urlpage)
```

```python
soup = BeautifulSoup(souce, 'html.parser')
```

```python
souce
```

you can view the html content by printing *soup

```python
print(soup.title.text)
```

*  All the *url link* connected to the main page are:

```python
for url in soup.find_all('a'):
    print(url.text)
```

```python
import pandas as pd
dfs = pd.read_html(urlpage, header = 0)
for df in dfs:
    print(df)
```

* length of the table

```python
table = soup.find('table', attrs={'class': 'tableSorter'})
results = table.find_all('tr')
print('Number of results', len(results))
```
