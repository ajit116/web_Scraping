# Web Scraping With BeautifulSoup

## Source:  

https://www.lseg.com/resources/1000-companies-inspire/2018-report-1000-companies-uk/search-1000-companies-uk-2018?results_per_page=1000

```python
#import dependencies
from bs4 import BeautifulSoup
import pandas as pd
import csv
from urllib.request import urlopen 
```

```python
url='https://www.lseg.com/resources/1000-companies-inspire/2018-report-1000-companies-uk/search-1000-companies-uk-2018?results_per_page=1000'
```

```python
souce = urlopen(url)
```

```python
soup = BeautifulSoup(souce,'html.parser')
```

### Title

```python
print(soup.title.text)
```

### Company Details in the class named panel-row-text

```python
container = soup.findAll('span', {'class': 'panel-row-text'})
```

### Details of first Company :

```python
for i in range(0,5):
    print(container[i])
    
```

```python
#lists
A=[]
B=[]
C=[]
D=[]
E=[]
```

```python
#storing Data in desired list
for i in range(0,len(container),5):
    A.append(container[i].text)
    B.append(container[i+1].text)
    C.append(container[i+2].text)
    D.append(container[i+3].text)
    E.append(container[i+4].text)
```

### Inserting list in the *Comp* Data frame

```python
Comp = pd.DataFrame(data = A, columns = ['Company name'])
Comp['Website']=B
Comp['Sector']=C
Comp['Region']=D
Comp['Revenue Band']=E
```
